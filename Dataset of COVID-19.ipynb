```r
library(ggplot2)
library(ggpubr)
```

```r
ggboxplot(data_1,y='TotalSamples',width=0.5)
```

```r
gghistogram(data_1,x='TotalSamples',bins = 10,add = "mean")
```

```r
ggplot(data_1,aes(x=TotalSamples,y=Positive))+geom_point(size=2.5,colour="red")+xlab("TotalSample")+ylab("Positive")+ggtitle("COVID-19 TESTING STATUS")
```

```r
ggplot(data_1,aes(x=State,y=Positive))+geom_bar(stat = 'Identity')+theme(axis.text.x = element_text(angle = 70,vjust = 0.5,colour='purple'))+xlab('State')+ylab('Positive')+ggtitle("STATEWISE_POSITIVE_STATUS")
```

```r
ggplot(data_1,aes(x=Date,y=TotalSamples))+geom_bar(stat = 'identity')+theme(axis.text.x = element_text(angle=70,vjust=0.5,colour='purple'))+xlab('Date')+ylab('ToatalSamples')+ggtitle("DATE_WISE_TOTAL_SAMPLE_CASE")
```

```r
ggplot(data_1,aes(x=Date,y=State))+geom_point(aes(colour=Date))+geom_smooth(method = 'lm')
```

```r
x<-geom_point(
    mapping=(aes(x=State,y=Date,colour=Date)),
    data=data_1,
    size=3
)
ggplot()+x
```
